```python
df_us_states = pd.read_csv('data/us-states.csv')
```

```python
df_us_states
```

```python
df_us_states = df_us_states.sort_values(["state", "date"]).reset_index()
```

```python
df_us_states
```

```python
# loop to reverse cumulative death count and get daily number of deaths
# it's a little tricky because we have to keep track of which
# state we're dealing with. The dataframe is ordered by state, so when we 
# encounter a new state, we need to reset the previous cumulative count to zero

daily_deaths = []
daily_cases = []
state = df_us_states.iloc[0]['state']
i = 0
daily_deaths.append(df_us_states.iloc[0]['deaths'])
daily_cases.append(df_us_states.iloc[0]['cases'])
pdc = daily_deaths[0]
pcc = daily_cases[0]
for d in df_us_states.iloc[1:].iterrows():
    
    if d[1]['state'] == state:
        daily_deaths.append(d[1]['deaths'] - pdc)
        daily_cases.append(d[1]['cases'] - pcc)
        pdc = d[1]['deaths']
        pcc = d[1]['cases']
    else:
        daily_deaths.append(d[1]['deaths'])
        daily_cases.append(d[1]['deaths'])
        pdc = 0
        pcc = 0
        state = d[1]['state']
    i += 1
```

```python
df_us_states['daily_deaths'] = daily_deaths
df_us_states['daily_cases'] = daily_cases
```
